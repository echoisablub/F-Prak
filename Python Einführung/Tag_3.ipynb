Das Argument `absolute_sigma`verändert offenkundig die Werte der Kovarianzmatrix.

````python
p, V =  opti.curve_fit(lambda x, m, a: m * x + a, U, I_sample[0], sigma=[Ierr]*len(I_sample[0]), absolute_sigma=True)
p2, V2 =  opti.curve_fit(lambda x, m, a: m * x + a, U, I_sample[0], sigma=[Ierr]*len(I_sample[0]), absolute_sigma=False)

print(p2/p)
print(V2/V)
````

Das Argument `absolute_sigma=False` führt offenbar zu einer Skalierung der Kovarianzmatrix. Das entspricht einer Skalierung des `sigma`der Messdaten mit der Wurzel des  Kovarianzskalierungsfaktors.

````python
p3, V3 =  opti.curve_fit(lambda x, m, a: m * x + a, U, I_sample[0], sigma=[Ierr*np.sqrt((V2/V)[0,0])]*len(I_sample[0]), absolute_sigma=True)
print(p3/p2)
print(V3/V2)
````

### 3.5 Fehler aus der inversen Hessematrix

Die Cramér-Rao-Schranke für die Varianz eines Schätzers ergibt: $\text{Var}_\theta \ge \left|\frac{1}{I(\theta)}\right|$.

Das $I$ steht hier für die Fisher-Information und diese kannn durch den Erwartungswert der zweiten Ableitung nach $\theta$ des Logarithmus der Wahrscheinlichkeitsdichte von $\theta$ berechnet werden $I(\theta) = \text{E}(\frac{\partial^2}{\partial \theta^2} \ln p(\theta))$.

In unserem Falle wäre also die Varianz $V(\hat \mu) = \left|\left(\frac{\partial^2 \ln L(\mu)}{\partial \mu^2}\right)^{-1}\right|$.

Die zweite Ableitung einer Funktion wird im mehrdimensionalen Fall die Hesse-Matrix genannt. Das Inverse der Hesse-Matrix ergibt also die Kovarianzmatrix des Schätzers.

### 4.3 Anpassungsgüte (goodness-of-fit)

Für den Test der Anpassungsgüte, also ob das Modell zu den Daten passt, wird überprüft, ob das Modell genauso "gut" funktioniert wie ein Modell mit maximal vielen Parametern. Dieses Modell wird *saturiertes Modell* genannt. In unserem Fall sind die beiden Hypothesen:
- die Daten $I_i$ werden von Normalverteilungen $G(\mu_i,\sigma_i)$ beschrieben mit $\mu_i = mU+a,\sigma_i = 0.5$
- die Daten $I_i$ werden von Normalverteilungen $G(\mu_i,\sigma_i)$ beschrieben mit $\mu_i = I_i,\sigma_i = 0.5$

Für die Entscheidung zwischen den beiden Hypothesen benutzen wir das Neyman-Pearson-Lemma, also ist das Verhältnis der Logarithmen der Wahrscheinlichkeitsdichten die beste Teststatistik.

Schreiben Sie eine Funktion

mit  $-\Delta \ln L(m,a) = -\ln L(m, a) + \ln L_\text{saturiert}= -\sum_i \ln G(I_i; \mu=mU+a,\sigma=0.5) + \sum_i \ln G(I_i; \mu=I_i,\sigma=0.5)$


Bestimmen Sie das Minimum von $-\Delta \ln L(m,a)$ und die Kovarianzmatrix. Welchen Wert hat $-\Delta \ln L(m,a)$ im Minimum? Teilen Sie diesen Wert durch den $\chi^2$-Wert von der Geradenanpassung aus Aufgabe 1.

### 4.4 Etwas Rechnen


Wenn man den Ausdruck ausrechnet, erhält man:
$$\begin{eqnarray}
-\Delta \ln L(m,a) &= &-\ln L(m, a) + \ln L_\text{saturiert}= -\sum_i \ln G(I_i; \mu=mU+a,\sigma=0.5) + \sum_i \ln G(I_i; \mu=I_i,\sigma=0.5) \\
& = & \sum_i \left(\ln G(I_i; \mu=I_i,\sigma=0.5) - \ln G(I_i; \mu=mU+a,\sigma=0.5)\right)\\
& = & \sum_i \left(\ln \frac{1}{\sqrt{2\pi}\sigma} - \frac{1}{2}\left(\frac{I_i-I_i}{\sigma}\right)^2 - \ln \frac{1}{\sqrt{2\pi}\sigma} + \frac{1}{2}\left(\frac{I_i-(mU+a)}{\sigma}\right)^2\right) \\
& = &  \sum_i \frac{1}{2}\left(\frac{I_i-(mU+a)}{\sigma}\right)^2 = \frac{1}{2} \chi^2
\end{eqnarray}$$

Es ist also sinnvoll besser für den Anpassunggütetest $-2\Delta \ln L$ zu betrachten, da dieses im Fall der Normalverteilung genau $\chi^2$-verteilt ist.
